```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#csv file path
transplant_data_file = "Data/Transplant___Transplant_Year_by_Recipient_Age_Organ.csv"
```

```python
#import csv, read it and make a dataframe
transplant_data_raw = pd.read_csv(transplant_data_file)
```

```python
#rename some columns
transplant_data_raw = transplant_data_raw.rename(columns={'Recipient Age': 'Recipient_Age', 
                                                          'Organ Type': 'Organ_Type', 'To Date': 'To_Date'})
transplant_data_raw
```

```python
#drop certain rows from transplant_data_raw from 'Recipient_Age' columns
rows_drop_list = ['Unknown']
transplant_data = transplant_data_raw[transplant_data_raw.Recipient_Age.isin(rows_drop_list) == False]
transplant_data
```

```python
#find list of all organ transplant types
transplant_organ_complete = transplant_data.Organ_Type.unique()
transplant_organ_complete
```

```python
#drop certain organ types from Organ_Type column
organs_drop_list = ['All Organs', 'Intestine', 'Abdominal Wall',
       'Head & Neck: Craniofacial', 'Head & Neck: Scalp',
       'Head & Neck: Larynx', 'GU: Penile', 'GU: Uterus',
       'Upper Limb: Bilateral', 'Upper Limb: Unilateral', 'Trachea']
transplant_data_new = transplant_data[transplant_data.Organ_Type.isin(organs_drop_list) == False]
transplant_data_new
```

```python
#drop nan values
transplant_data_drop = transplant_data_new.dropna()
transplant_data_drop
```

```python
#get list of all recipient age groups
recipient_age_groups = transplant_data_drop.Recipient_Age.unique()
recipient_age_groups
```

```python
#gather just the pediatric data
pediatric_ages = ['< 1 Year', '1-5 Years', '6-10 Years',
       '11-17 Years']
transplant_data_peds = transplant_data_drop[transplant_data_drop['Recipient_Age'].isin(pediatric_ages)]
transplant_data_peds
```

```python
#calculate years to adulthood. Create variable calculations. Under a year is +18, 1-5 is +17, 6-10 is +12, 
transplant_data_peds.to_csv('file_name.csv')
```
